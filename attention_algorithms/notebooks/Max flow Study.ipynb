# Study of the maximum flow problem in the attention graph (a more complex approach to the problem)

---

The objective of this notebook is to study the *maximum flow* problem in the attention graph. The structure of the attention graph is given into the file .\attention_algorithms\raw_attention.py

The attention graph is based on the heads aggregation over the attention_tensor, for now we have to main implementation for the head aggregation :
- the maximum head aggregation
- the mean head aggregation

## Load the environnement and the model

```python
%load_ext autoreload
%autoreload 2
import os
import matplotlib.pyplot as plt
import pandas as pd
import torch

from os import path
import tqdm

from training_bert import BertNliLight
from custom_data_set import SnliDataset
from custom_data_set import test_dir
from torch.utils.data import DataLoader

from attention_algorithms.raw_attention import RawAttention
from attention_algorithms.attention_flow import attention_flow_max

from attention_algorithms.raw_attention import normalize_attention, softmax_normalization
from attention_algorithms.attention_metrics import plot_roc_curve, combine_roc_curves

# set the repository to the git repository
cwd = os.getcwd().split(os.path.sep)
while cwd[-1] != "stage_4_gm":
    os.chdir("..")
    cwd = os.getcwd().split(os.path.sep)
print(">> the git rep : ", end="")
print(os.getcwd())

# the folder where we will save our data
plots_folder = os.path.join(os.getcwd(), '.cache', 'plots')
graph_folder = path.join(plots_folder, "max_flow_approach")
if not path.exists(path.join(plots_folder, "max_flow_approach")):
    os.mkdir(path.join(plots_folder, "max_flow_approach"))

print(f">> the plots location : {graph_folder}")
```

```python
# load the model

ckp = path.join(".cache", "logs", "igrida_trained", "0", "best.ckpt")
model = BertNliLight.load_from_checkpoint(ckp)
model.eval()  # make sure to have the model in eval mod before using it
```

## Calculation of the attention values for the test set

```python
# load the data
# load some data just load one sentence
data_set = SnliDataset(dir=test_dir, nb_sentences=500, msg=False)
data_loader = DataLoader(data_set, batch_size=500, shuffle=False)

sentences, masks, train_labels = next(iter(data_loader))

print(f">> shape of the sentences : {sentences.shape}")
print(f">> shape of the masks : {masks.shape}")
```

```python
# the data we annotated.
e_snli_data = pd.read_csv(os.path.join('.cache', 'raw_data', 'e_snli', 'cleaned_data', 'test.csv'), sep=",")\
    [["tok_sent", "hg_goal"]]
e_snli_data.head()
```

Now for each sentence in the snli dataset we will calculate in the attention graph the flow max from the token CLS to the input tokens. The attention graph will be constructed by two means. First we will construct it with the maximum head aggregation, then with the mean head aggregation

```python
class LenException(Exception):
    pass

Y_test = []
probs_avg = []
probs_max = []

nb_err = 0

for _, i in enumerate(tqdm.tqdm(range(500))):
    j = 0
    # iteration through all the sentences
    # construct the raw attention object
    sent = sentences[i, :].clone().detach()[None,:]
    mk = masks[i, :].clone().detach()[None,:]
    raw_attention_inst = RawAttention(model=model,
                                      input_ids=sent,
                                      attention_mask=mk,
                                      test_mod=False
                                      )
    try :
        # >> search for the corresponding sentence in the e-snli dataset
        # >> it can happen that we do not find the right sentence we just skip and go to the next sentence
        while j<e_snli_data.shape[0] and raw_attention_inst.tokens!=eval(e_snli_data["tok_sent"][j]):
            j+=1

        if j>=e_snli_data.shape[0]:
            raise LenException
        else :
            Y_test += eval(e_snli_data["hg_goal"][j])

            # maximum agregation for the heads
            raw_attention_inst.set_up_graph(agr_type="max")
            flow = attention_flow_max(raw_attention_inst)
            flow = normalize_attention(raw_attention_inst.tokens , torch.tensor(flow))
            probs_max += flow

            # avg agregation for the heads
            raw_attention_inst.set_up_graph(agr_type="avg")
            flow = attention_flow_max(raw_attention_inst)
            flow = normalize_attention(raw_attention_inst.tokens , torch.tensor(flow))
            probs_avg += flow


    except LenException :
        # count the different errors
        nb_err += 1



print(f">> nb_errors {nb_err}")

print(f">> len Y_test {len(Y_test)}")
print(f">> len probs_avg {len(probs_avg)}")
print(f">> len probs_avg {len(probs_max)}")
```

## The roc curves

### "Max" head aggregation

```python
fig, best_tr = plot_roc_curve(Y_test , probs_max)
plt.savefig(os.path.join(graph_folder , "max_agreg_test_set_roc_curve.png"))
```

### "Avg" head aggregation

```python
fig, best_tr = plot_roc_curve(Y_test , probs_avg)
plt.savefig(os.path.join(graph_folder , "avg_agreg_test_set_roc_curve.png"))
```

### Combination

```python
probs = [probs_max, probs_avg]

fig = combine_roc_curves(Y_test, probs)
plt.savefig(os.path.join(graph_folder, "combined_roc_curves.png"))
```

## Flow max from different layers
---
In our problem we computed the flow max from the *[CLS]* token from the last layer (layer 12) to the input of the model. We want to test now what happens if we compute this maximum flow between the *[CLS]* token and the input but if we take this token on an intermediate layer.

### "Max" head agregation

```python
# TODO : check this cell.
Y_test = []
probs = []

nb_err = 0

# here only on 50 sentences
for _, i in enumerate(tqdm.tqdm(range(50))):
    j = 0
    # iteration through all the sentences
    # construct the raw attention object
    sent = sentences[i, :].clone().detach()[None,:]
    mk = masks[i, :].clone().detach()[None,:]
    raw_attention_inst = RawAttention(model=model,
                                      input_ids=sent,
                                      attention_mask=mk,
                                      test_mod=False
                                      )
    try :
        while j<e_snli_data.shape[0] and raw_attention_inst.tokens!=eval(e_snli_data["tok_sent"][j]):
            j+=1

        if j>=e_snli_data.shape[0]:
            raise LenException
        else :
            Y_test += eval(e_snli_data["hg_goal"][j])

            # maximum agregation for the heads
            raw_attention_inst.set_up_graph(agr_type="max")
            n_layer, _, _ = raw_attention_inst.att_tens_agr.shape
            for l in range(n_layer+1):
                if l>0:
                    flow = attention_flow_max(raw_attention_inst, out_layer=l)
                    flow = normalize_attention(raw_attention_inst.tokens , torch.tensor(flow))
                    if len(probs)==l-1:
                        probs.append(list(flow.detach().numpy()))
                    else :
                        probs[l-1]+=list(flow.detach().numpy())


    except LenException :
        # count the different errors
        nb_err += 1



print(f">> nb_errors {nb_err}")

print(f">> len Y_test {len(Y_test)}")
print(f">> len probs_avg {len(probs)}")
print(f">> len probs_avg {len(probs[0])}")
```

```python
legend = [f"layer {i}" for i in range(1,13)]
fig = combine_roc_curves(Y_test, probs, legend=legend)
plt.savefig(os.path.join(graph_folder, "every_layer_combined_roc_curves.png"))
```
